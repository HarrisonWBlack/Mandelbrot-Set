```
import numpy
from numba import autojit
import matplotlib.pyplot as plt
```

```
def mandelbrot(real, imaginary, max_iteration):
  ''' Return the value for the current step in the set '''
  
  c = complex(real, imaginary)
  z = 0.0j
  
  for i in range(max_iteration):
    z = z * z + c
    if(z.real * z.real + z.imag * z.imag) >= 4:
      return i
    
  return max_iteration
```

```
cols = 3000
rows = 3000

result = numpy.zeros([rows, cols])
for row_index, real in enumerate(numpy.linspace(-2, 1, num=rows)):
  for col_index, imaginary in enumerate(numpy.linspace(-1, 1, num=cols)):
    result[row_index, col_index] = mandelbrot(real, imaginary, 100)
```

```
# Increase dpi to make larger
plt.figure(dpi=200)
plt.imshow(result.T, cmap='jet', interpolation='bilinear', extent=[-2, 1, -1, 1])
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.axis('off')
plt.show()
```
